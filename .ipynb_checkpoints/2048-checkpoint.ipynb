```python
import random

class Matrix:
    def __init__(self,row,col,fill=0):
        self.row = row
        self.col = col
        self.fill = fill
        self.matrix = []
        for i in range(row):
            r = []
            for j in range(col):
                r.append(fill)
            self.matrix.append(r)
            
    def get(self,row,col):
        if(self.row > row and self.col > col):
            return self.matrix[row][col]
        else:
            return -1
        
    def set(self,row,col,item):
        if(self.row > row and self.col > col):
            self.matrix[row][col] = item
            return 0
        else:
            return -1
    
    def full(self):
        for i in range(self.row):
            for j in range(self.col):
                if(self.get(i,j) == self.fill):
                    return False
        return True
        
    def newBlock(self,item):
        done = False
        while done == False:
            row = random.randint(-1,self.row)
            col = random.randint(-1,self.col)
            if(self.get(row,col) == self.fill):
                done = True
                self.set(row,col,item)
    

        
    def testOver(self):
        for i in range(self.row):
            for j in range(self.col-1):
                if(self.get(i,j) == self.get(i,j+1)):
                    return False
        for i in range(self.row-1):
            for j in range(self.col):
                if(self.get(i,j) == self.get(i+1,j)):
                    return False
        return True
    

    # direction table: 0 = left, 1 = down, 2 = right, 3 = up
    def move_fill(self,direction):
        if(direction == 0):
            for i in range(self.row):
                for j in range(self.col-1,-1,-1):
                    item = self.get(i,j)
                    if(item != self.fill):
                        for k in range(j,-1,-1):
                            if(self.get(i,k) == self.fill):
                                self.set(i,k,item)
                                self.set(i,j,self.fill)
                                break
                            
        elif(direction == 1):
            for i in range(self.row-1,-1,-1):
                for j in range(self.col):
                    item = self.get(i,j)
                    if(item != self.fill):

                        for k in range(i,-1,-1):
                            if(self.get(k,j) == self.fill):
                                self.set(i,k,item)
                                self.set(i,j,self.fill)
                                break

        elif(direction == 2):
            for i in range(self.row):
                for j in range(self.col):
                    item = self.get(i,j)
                    if(item != self.fill):
                        for k in range(j,self.col):
                            if(self.get(i,k) == self.fill):
                                self.set(i,k,item)
                                self.set(i,j,self.fill)
                                break
                                
        elif(direction == 3):
            for i in range(self.row):
                for j in range(self.col):
                    item = self.get(i,j)
                    if(item != self.fill):
                        for k in range(i,self.row):
                            if(self.get(k,j) == self.fill):
                                self.set(i,k,item)
                                self.set(i,j,self.fill)
                                break
        else:
            return -1     
        
    
    def printBoard(self):
        for i in range(self.row):
            for j in range(self.col):
                print(self.get(i,j), end='')
            print(' ')
        print(' ')    
            
            
            
mt = Matrix(3,3)
mt.printBoard()

mt.set(0,2,4)


mt.printBoard()
mt.move_fill(1)
mt.printBoard()
mt.move_fill(2)
mt.printBoard()
mt.move_fill(3)
mt.printBoard()

```
